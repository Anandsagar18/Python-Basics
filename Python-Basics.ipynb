# Python Basics

### Theory

Q1. What is Python, and why is it popular?
- Python is a Programming language and it is popular because of it's vast community base and industry acceptance in the field of Data Science.

Q2. What is an interpreter in Python?
- Interpreter is a program that executes Python code line by line in real time.

Q3. What are pre-defined keywords in Python?
- These are words in python having special meaning and pre-defined functions in python language.

Q4. Can keywords be used as variable names?
- No, these cannot be used as identifiers as they serve specific purpose in python program.

Q5. What is mutability in Python?
- Mutability means the ability to change or update the contents of a variable or data structure.

Q6. Why are lists mutable, but tuples are immutable?
- They are designed as such in Python. So, lists for dynamic data and tuple for static data.

Q7. What is the difference between “==” and “is” operators in Python?
- "==" compares the value of the variables whereas "is" compares the location of the variables.

Q8. What are logical operators in Python?
- "and", "or" and "not" are logical operators in python they work as logic gates and return bool value ie True/False.

Q9. What is type casting in Python?
- Type casting is basically changing the data type of a variable to another one.

Q10. What is the difference between implicit and explicit type casting?
- implicit means when python automatically does type casting to perform some operations on the other hand explicit means manual type casting by the programmer.

Q11. What is the purpose of conditional statements in Python?
- To control the flow of the program and they allow us to make decisions.

Q12. How does the elif statement work?
- It means else-f, we check this only if the previous "if" or "elif" statements are found to be false.

Q13. What is the difference between for and while loops?
- We generally use for loop for iterating over data structures like strin, lists, tuples, etc whereas we use while loop to perform same operation n-number of times.

Q14. Describe a scenario where a while loop is more suitable than a for loop.
- when the number of repetitions is not known before hand while loop is better than a for loop.

```python
#6 Write a function to demonstrate the behavior of mutable and immutable arguments.
def modify_list(lst):
    lst.append(100)
def modify_string(s):
    s += " World"
my_list = [1, 2, 3]
my_string = "Hello"
modify_list(my_list)
modify_string(my_string)
print(f"Modified list: {my_list}")  # Output: [1, 2, 3, 100]
print(f"Modified string: {my_string}")  # Output: "Hello"
```

```python
#7 Write a program that performs basic arithmetic operations on two user-input numbers.
first_num = int(input("Enter first number: "))
second_num = int(input("Enter second number: "))
addition = first_num + second_num
subtraction = first_num - second_num
multiplication = first_num * second_num
division = first_num / second_num
modulo = first_num % second_num
quotient = first_num // second_num
exponentiation = first_num ** second_num
print(f"Addition: {addition}")
print(f"Subtraction: {subtraction}")
print(f"Multiplication: {multiplication}")
print(f"Division: {division}")
print(f"Modulo: {modulo}")
print(f"Quotient: {quotient}")
print(f"Exponentiation: {exponentiation}")
```

```python
#13  Write a Python program to find the sum of all even numbers between 1 and 50.
sum = 0
for i in range(0, 51, 2):
    sum += i
print(f"The sum of all even numbers between 1 and 50 is: {sum}")
```

```python
#14 Write a program to reverse a string using a while loop.
string = input("Enter a string: ")
reversed_string = ""
length = len(string)
while length > 0:
    reversed_string += string[length - 1]
    length -= 1
print(f"Reversed string: {reversed_string}")
```

```python
#15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number to calculate its factorial: "))
fact =  1
while num>0:
    fact *=num
    num -=1
print(f"The factorial is: {fact}")
```
